```python
# 문제 4
print("문제4: 각 열의 결측치 확인")
print(df.isnull().sum())
```

```python
# 문제 5

outliers = df[(df['점수'] > 100) | (df['공부시간'] > 10)]

print("문제5: 이상치 확인 (조건식)")
print(f"이상치 개수: {len(outliers)}")
print("이상치 데이터:")
outliers
```

```python
# 중복 데이터는 완전히 똑같은 데이터

duplicate_1 = df[df.duplicated(keep=False)]



print("문제 6: 완전 중복 행 확인")
print(f"중복 제거 전 데이터프레임 크기: {df.shape}")
print(f"완전 중복 행 개수: {len(duplicate_1)}")
duplicate_1
```

```python
# 문제 7: 중복 데이터 삭제하기
# 데이터프레임에서 중복된 데이터를 삭제하고, 중복이 제거된 데이터프레임을 출력하세요

duplicate_2 = df[df.duplicated()]
df_1 = df.drop_duplicates()

print("문제7: 완전 중복 제거")
print(f"중복 제거 후 데이터프레임 크기: {df_1.shape}")
print(f"제거된 행 수: {len(duplicate_2)}")

df_1
```
