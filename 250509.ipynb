```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
plt.ylim(0, 400)

Month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

# 그래프 그리기
plt.plot(Month, Rainfall, 'bo-.')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Rainfall(mm)", fontsize=14)
plt.title("Monthly Rainfall", fontsize=18)

# annotate를 사용하여 각 점 위에 강수량 표시
for i in range(len(Rainfall)):
    text = str(Rainfall[i])
    plt.annotate(f"{text}mm",
                 (Month[i], Rainfall[i]),
                 textcoords="offset points",
                 xytext=(0, 10),  # 텍스트를 점 위에 10 포인트 위치시킴
                 )

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
plt.ylim(0, 400)

Month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

# 그래프 그리기
plt.plot(Month, Rainfall, 'bo-.')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Rainfall(mm)", fontsize=14)
plt.title("Monthly Rainfall", fontsize=18)

# text()를 사용하여 각 점 위에 강수량 표시
for i in range(len(Rainfall)):
    text = str(Rainfall[i])
    plt.text(Month[i], Rainfall[i] + 10, f"{text}mm")

plt.show()
```
